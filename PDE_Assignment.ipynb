```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-30,30,.1)
x4 = np.arange(1,10,.1)
x5 = np.arange(1,10,.1)
fig, ax = plt.subplots()
a1 = x+1
a2 = -x+1
a3 = x-1
a5 = -x-1
a34 = x4-1
a14 = x4+1
a35 = x5-1
plt.plot(x, a1,
         x, a2,
         x, a3,
         x, a5)
plt.ylim(0, 30)

plt.fill_between(x4, a14, a34, color='green', alpha=1)
plt.fill_between(x5, a35, color='red', alpha=1)

plt.show()
```
